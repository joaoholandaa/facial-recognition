```python
image = cv.imread("got.jpg")
```

```python
type(image)
```

```python
def show_image(image):
  image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  plt.imshow(image_rgb)
show_image(image)
```

```python
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
show_image(image_gray)
```

```python
faceCascade = cv.CascadeClassifier(cv.data.haarcascades +
                                   "haarcascade_frontalface_default.xml")
```

```python
faces = faceCascade.detectMultiScale(image_gray)
```

```python
image_copy = np.array(image)
```

```python
for x, y, w, h in faces:
  cv.rectangle(image_copy, (x,y), (x+w, y+h), (0,255,0), 2)
show_image(image_copy)
```

```python
images_cropped = list()
```

```python
for x, y, w, h in faces:
  face = image[y:y+h, x:x+w]
  face = cv.resize(face, (160,160))
  images_cropped.append(face)
```
